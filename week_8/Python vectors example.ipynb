```python
# Matrices
```

```python
r = np.array([3,4])
plot_vector([r])
```

```python
M = np.array([[0,-1],
              [1,0]])
```

```python
plot_linear_transformation(M)
```

```python
r_prime = np.matmul(M,r)
print(r_prime)
```

```python
plot_vector([r_prime])
```

```python
# computing the inverse of a matrix
N = np.linalg.inv(M)
print(N)
```

```python
# check this is a 90 degree clockwise rotation
plot_linear_transformation(N)
```

```python
q = np.matmul(N, r_prime)
# what is q?
```

```python
print(q)
plot_vector([q])
# It's our original vector r!!
```

```python
# Matrix-matrix multiplication
O = np.matmul(M,M)
print(O)
plot_linear_transformation(O)
```

```python
s = np.matmul(O,r)
plot_vector([s])
print(s)
```

```python
# What happens if we multiply a matrix by its inverse?
```

```python
M_inv = np.linalg.inv(M)
plot_linear_transformation(M_inv)
```

```python
S = np.matmul(M, M_inv)
print(S)
```

```python
plot_linear_transformation(S)
```

```python
# Nothing has happened! 
# A matrix times its inverse gives the "identity" matrix, which leaves i and j unchanged.
```
